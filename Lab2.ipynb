```python
import math
import numpy as np
import csv
#R=20
#r=13
#rho=7

x=[]
y=[]
def points(R,r,rho):
    '''
    Domain: R,r,rho (floats) and list_1 which is list of number of data points.
    Return list of x and y coordinates
    '''
    list_1=np.linspace(0,math.pi*2*r,2000)
    g_code=[]
    for t in list_1:
        val_x=(R-r)*math.cos(t)+rho*math.cos((R-r)*t/r)+90
        x.append(val_x)
        val_y=(R-r)*math.sin(t)-rho*math.sin((R-r)*t/r)+90
        y.append(val_y)
        setcoord = lambda x,y: f'G0 X{str(x)} Y{str(y)}\n'
        g_code.append(setcoord(val_x,val_y))
    return g_code
    

def wr(R,r,rho):
    '''
    take list from previews fuctionand write gcode file
    '''
    codes=points(R,r,rho)
    with open('g_code_1.gcode','w') as f:
        f.write('G28 ; home the axes\n')
        f.write('G0 Z10 ; move the pen up above the plate\n')
        f.write('G0 X50 Y50 ; move the pen above the corner of the build plate\n')
        f.write('G0 Z0 ; move the pen down onto the paper\n')
        

        for line in codes:
            f.write(line)
            #f.write('\n')
            

wr(20,13,7)
   

import math
import random
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
'''
plot graph
'''


plt.plot(x, y)
plt.show()
    
        
    
```

```python
import math
import numpy as np
#R1=20
#r1=13
#rho1=7

x1=[]
y1=[]
def points_1(R1,r1,rho1):
    '''
    domain: floats, list
    codomain: lists of x,y coordinate
    '''
    list_2=np.linspace(0,math.pi*2*r1,2000)
    eg_x=map(lambda t:(R1-r1)*math.cos(t)+rho1*math.cos((R1-r1)*t/r1)+90, list_2)
    eg_y=map(lambda t:(R1-r1)*math.sin(t)-rho1*math.sin((R1-r1)*t/r1)+90, list_2)
    setcoord = map(lambda x,y: f'G0 X{str(x)} Y{str(y)}\n',x,y)
    g_code=list(setcoord)
    x1.extend(list(eg_x))
    y1.extend(list(eg_y))
    return g_code



def wr(R1,r1,rho1):
    '''
    use lists from points_1 and write gcode
    '''
    codes=points_1(R1,r1,rho1)
    with open('g_code_1.gcode','w') as f:
        f.write('G28 ; home the axes\n')
        f.write('G0 Z10 ; move the pen up above the plate\n')
        f.write('G0 X50 Y50 ; move the pen above the corner of the build plate\n')
        f.write('G0 Z0 ; move the pen down onto the paper\n')
        
        for line in codes:
            f.write(line)
            
            

wr(20,13,7)



import math
import random
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.plot(x1, y1)
plt.show()
```
